# Velocity Image Editor

```python
# Libraries
import numpy as np
from matplotlib import pyplot as plt
```

```python
file_name = 'seismogram_00800_str160.bin'
file_name_output = 'seismogram_00800_str160.png'

dimension_x = 320
dimension_z = 1201

spacing_x = 25.0
spacing_z = 4

scaling = 1.0/1000.0
normalization = 10.0/1.0
```

```python
# Adjusting the data
data = np.fromfile(file_name,dtype='float32')
data = data.reshape(dimension_x,dimension_z)
data = data.transpose()

data = data * normalization
```

```python
# Scaling the axis to the original dimension
xmin = (0 * spacing_x) * scaling
xmax = (dimension_x * spacing_x) * scaling
zmin = (0 * spacing_z) * scaling
zmax = (dimension_z * spacing_z) * scaling
```

```python
# setting xtick on top
plt.rcParams['xtick.bottom'] = False
plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = True
plt.rcParams['xtick.labeltop'] = True

plt.rc('xtick', labelsize=14)     
plt.rc('ytick', labelsize=14)
plt.rcParams.update({'font.size': 14}) 

# Setting figure
fig, ax = plt.subplots(figsize=(8,10))
ax.xaxis.set_label_position('top') # moving extension to the top
edit_fig = ax.imshow(data, cmap='Greys', aspect='auto', vmin=-1.0, vmax=1.0, interpolation='bicubic', extent=[xmin,xmax,zmax,zmin])
cbar = fig.colorbar(edit_fig, ax=ax, fraction=0.18, pad=0.02, label='Amplitude', aspect=30, format='%.0e')
plt.xlabel('Offset (Km)', fontsize=14)
plt.ylabel('Time (s)', fontsize=14)
#plt.text(0.55, 0.15, "A", size=16, rotation=0.,
#         ha="center", va="center",
#         bbox=dict(boxstyle="round", facecolor='white'))

plt.savefig(file_name_output, dpi=300, bbox_inches='tight')
plt.show()
```
