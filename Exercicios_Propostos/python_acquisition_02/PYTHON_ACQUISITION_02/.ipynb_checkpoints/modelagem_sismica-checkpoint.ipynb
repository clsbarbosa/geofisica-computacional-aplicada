```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
from numba import jit
```

```python
dimX = 300
dimZ = 200

vp = np.zeros((dimZ, dimX))

for k in range(dimX):
    for i in range(int(dimZ/3)):
        vp[i, k] = 1500
    for i in range(int(dimZ/3), int(dimZ)):
        vp[i, k] = 2000

plt.imshow(vp)
```

```python
Na = 0 # Número de pontos da borda
Nzz = dimZ + 2 * Na
Nxx = dimX + 2 * Na
```

```python
C = np.zeros((Nzz, Nxx))                # Matriz para simplificação da equação da onda discretizada
A = np.zeros((Nzz, Nxx))                # Matriz para simplificação da equação de bordas não reflexivas

h = 10
dt = 0.001     # Variação temporal

extended_vp = np.copy(vp)

for i in range(Nzz):
    for k in range(Nxx):
        A[i, k] = extended_vp[i, k] * (dt/h)
        C[i, k] = - (A[i, k] ** 2) / 12
```

```python
fcorte = 30    # frequência de corte
dt = 0.001     # Variação temporal

Nf = int(4 * sqrt(pi) / (fcorte * dt))

fonte = np.zeros(Nf)
t = np.zeros(Nf)

TF = 2 * sqrt(pi) / fcorte       # Período da função Gaussiana
fc = fcorte / (3. * sqrt(pi))    # Frequência central


for n in range(Nf):
    t[n] = ((n) * dt - TF)
    fonte[n] = -(-exp(-pi * (pi * fc * t[n]) ** 2) * (1 - 2 * pi * (pi * fc * t[n]) * (pi * fc * t[n])))

plt.grid()
plt.plot(t, fonte)
```

```python
@jit  # Compilador Just-in-time
def modelagem(P1, P2, P3, C, A, sis, fonte, Nf, init_shot_x, init_shot_z, ntotal, rec_shotz):
    
    for n in range(0,ntotal,1):

        # Cálculo do Campo no interior do modelo
        for i in range(2, Nzz-2):
            for k in range(2, Nxx-2):

                P3[i, k] = C[i, k] * (P2[i, k + 2] + P2[i, k - 2] + P2[i + 2, k] + P2[i - 2, k] - 16 * (P2[i, k + 1] + P2[i, k - 1] + P2[i + 1, k] + P2[i - 1, k]) + 60 * (P2[i, k])) + 2 * (P2[i, k]) - P1[i, k]

        # Termo fonte
        if n < Nf:
            v = A[init_shot_z, init_shot_x]
            
            P2[init_shot_z, init_shot_x] = P2[init_shot_z, init_shot_x] + (fonte[n] * v*v)
            
        for k in range (dimX):
            sis[n, k] = P3[rec_shotz,k+Na]

        # Atualização do campo de onda
        P1 = np.copy(P2)
        P2 = np.copy(P3)

        if n % 1000 == 0:
            print(n)

    return sis
```

```python
rec_shotz = 5 + Na
init_shot_z = 5 + Na
init_shot_x = 150 + Na
ntotal = 2000

P1 = np.zeros((Nzz, Nxx))               # Matriz no tempo passado
P2 = np.zeros((Nzz, Nxx))               # Matriz no tempo presente
P3 = np.zeros((Nzz, Nxx))               # Matriz no tempo futuro
sis = np.zeros((ntotal, dimX))            # Matriz do sismograma
```

```python
sis = modelagem(P1, P2, P3, C, A, sis, fonte, Nf, init_shot_x, init_shot_z, ntotal, rec_shotz)
```

```python
# Rotina para exbição do sismograma gerado
def plot_sis(sis, name):

    ## Eixos em dimensões reais
    scaleX = 1.0 / 1000
    scaleZ = 1

    Xmin = (0 * h) * scaleX
    Xmax = (Nxx * h) * scaleX
    Zmin = (0 * dt) * scaleZ
    Zmax = (ntotal * dt) * scaleZ

    data= sis / np.max(np.abs(sis))   # nomalizando o sismograma (sis)

    ## Eixo horizontal no topo da figura
    plt.rcParams['xtick.bottom'] = False
    plt.rcParams['xtick.labelbottom'] = False
    plt.rcParams['xtick.top'] = True
    plt.rcParams['xtick.labeltop'] = True

    plt.rc('xtick', labelsize=16)
    plt.rc('ytick', labelsize=16)
    plt.rcParams.update({'font.size':16})

    ## Formatação do plot
    fig, ax = plt.subplots(figsize=(12,12))
    ax.xaxis.set_label_position('top')
    im = ax.imshow(data, cmap='gray', vmin=-0.05,vmax=0.05,extent=[Xmin,Xmax,Zmax,Zmin], interpolation='bicubic', aspect=2)
    cbar = fig.colorbar(im, ax=ax, fraction=0.04, pad=0.08, label='Amplitude', aspect=15)
#    plt.title('Marmousi')
    plt.xlabel('Distance (Km)', fontsize=16)
    plt.ylabel('Time (s)', fontsize=16)
    plt.savefig(name, dpi=300,bbox_inches='tight')

    plt.show()
```

```python
plot_sis(sis,'sismogram.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
from numba import jit                         
```

```python

h = 10.0                                       # espaçamento da malha
dt = 0.001                                      # taxa de variação temporal
dimX = 300                                    # extensão do domínio
dimZ = 200                                    # profundidade do domínio
Ntotal = 2000                                # tempo total discreto
```

```python
fonte       = np.zeros(Ntotal)          # fonte sísmica
velocity    = np.zeros((dimZ, dimX))      # campo de velocidades
C           = np.zeros((dimZ, dimX))      # matriz auxilar
    
wavefield01 = np.zeros((dimZ, dimX))      # campo de pressão em n - dt
wavefield02 = np.zeros((dimZ, dimX))      # campo de pressão em n
wavefield03 = np.zeros((dimZ, dimX))      # campo de pressão em n + dt

seismic_signal = np.zeros(Ntotal)       # sinal sísmico em 1 receptor

sis = np.zeros((Ntotal, dimX))            # Matriz do sismograma
```

```python
# posicionamento fonte sísmica
shot_x = 150
shot_z = 5

# posicionamento receptores
rec_x = 150
rec_z = 150                           
```

```python
fcorte = 30.0    # frequência de corte

Nf = int(4 * sqrt(pi) / (fcorte * dt))

t = np.zeros(Ntotal)

TF = 2 * sqrt(pi) / fcorte       # Período da função Gaussiana
fc = fcorte / (3. * sqrt(pi))    # Frequência central


for n in range(Ntotal):
    t[n] = ((n) * dt - TF)
    fonte[n] = -(-exp(-pi * (pi * fc * t[n]) ** 2) * (1 - 2 * pi * (pi * fc * t[n]) * (pi * fc * t[n])))

plt.grid()
plt.plot(t, fonte)
```

```python
for i in range(dimZ):
    for k in range(dimX):
        C[i, k] = (velocity[i, k] * (dt/h)) ** 2
        
c = (-2.0, 1.0)
```

```python
for n in range(0, Ntotal):

    #kernel da equação da onda sísmica
    for i in range(2, dimZ - 2):
        for k in range(2, dimX - 2):
            wavefield03[i,k] = C[i,k] * (2.0*c[0]*wavefield02[i, k] + c[1]*(wavefield02[i,k-1] + wavefield02[i+1,k]+ wavefield02[i-1,k] + wavefield02[i+1,k])) + 2.0*wavefield02[i,k] - wavefield01[i,k]

    wavefield02[shot_z,shot_x] = wavefield02[shot_z,shot_x] + fonte[n]*(velocity[shot_z,shot_x]**2)
    
    seismic_signal[n] = wavefield02[rec_z,rec_x]
    
    for k in range (dimX):
            sis[n, k] = wavefield03[rec_z,k]
            
    # Atualização do campo de onda
    wavefield01 = np.copy(wavefield02)
    wavefield02 = np.copy(wavefield03)
```
